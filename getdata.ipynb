```python
import pandas as pd
import numpy as np
import requests
import time

with open('sports_apikey.txt', 'r') as file:
    api_key = file.read()
```

```python
## Many urls used because api limits to 500 games per query
urls =["https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2021-01-01",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2021-03-24",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2021-06-11",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2021-12-02",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2022-02-09",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2022-05-20",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2022-12-04",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2023-02-12",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2023-07-09",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2023-12-17",
    "https://api.apilayer.com/therundown/sports/4/schedule?limit=1000&from=2024-03-01"]

# Intialize a place for the data to go
all_schedules = []
payload = {}

# Load in API Key
headers= {
  "apikey": api_key
}


# Run for loop
for url in urls:
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        all_schedules.extend(data['schedules'])
    else:
        print(f"Error fetching data from {url}: {response.status_code}")

    time.sleep(5) # sleep for 5 seconds to follow ethical practices
```

```python
# Create a DataFrame from the combined data
df = pd.DataFrame(all_schedules)
df
```

```python
# Write the file
df.to_csv('sports_data.csv')
```
